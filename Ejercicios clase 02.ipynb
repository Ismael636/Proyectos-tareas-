```
#conversiones de variables

y ="2"
print(type(y),y)
s =int(y) #se utiliza la abreviatura de la variable que queremos convertir
#las variables son entero(int), flotante(float), texto(str), booleanos(bool)
print(type(s),s)
x =str(y)
print(type(x),x)
```
