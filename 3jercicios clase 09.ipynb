```
#Modulo numpy

import numpy as nombre #es muy comun escribir np
a = [1,2,3]
print(type(a))
a = nombre.array(a)# hace el cambio a tipo vector
print(type(a))
print(3*a)
print(a+a)
```

```
diez = nombre.linspace(1,10,10)
print(diez)

ceros = nombre.zeros(3)
print(ceros)
```

```
b = nombre.array([-1,-60,12,3.5,2,3,3])
Min = nombre.amin(b) # la funcion amin da el valor mas peque√±o de un vector
print(Min)

Max = nombre.amax(b)#la funcion max da el valor mas grande
print(Max)

donde = nombre.where(b == 3.5)#indica en que posicion esta un valor del vector
print(donde)
donde2 = nombre.where(b==3)
print(donde2)
```

```
#x^2 - 3x + 2 = 0, ecuacion a resolver

ec1 = nombre.array([1,-3,2])#se utiliza como la chicharronera
solu1 = nombre.roots(ec1)
print(solu1)
print('\n')
#4x^4 - 2x^2 + 3x -2 = 0
ec2 = nombre.array([4,-2,3,-2])
solu2 = nombre.roots(ec2)
print(solu2)
```

```
#5x - 3y -z = 1
#x + 4y -6z = -1
#2x + 3y + 4x = 9
X = nombre.array([[5,-3,-1],[1,4,-6],[2,3,4]])
val = nombre.array([1,-1,9])
print(X)
print(val)
print('\n')
solucion = nombre.linalg.solve(X,val)#resuelve un sistema de ecuaciones
print(solucion)
```

```
y = nombre.array([[2,-6,-1],[9,3,-6],[7,3,7]])
#print(y)
XY = nombre.matmul(X,y)
print(XY)
```
