```
lista = ['[0,1]\n','[2,3]\n','[4,1]\n','[9,6]']
with open('coordenadas.txt','w') as file:
  file.writelines(lista)
```

```
with open('coordenadas.txt','r') as coord:
  lineas = [linea.strip() for linea in coord]

  for linea in lineas:
    print(linea)

coord.close()
```
