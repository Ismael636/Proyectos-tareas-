```
#operadores de asignacion

A=5
#A=A+3
#print(A)
A+=3#Es equivalente a escribir A=A+3
print(A)

#resta de asignacion
A-=2
print(A)
```

```
#multiplicacion de asignacion
A*=2#Es equivalente a escribir A=2*A
print(A)

#DIvision de asignacion
A/=4 #Equivale a escribir A=A/4
print(A)
```

```
#Potencia en asignacion
A**=3#Equivale a escribir A^3
print(A)
```

```
#Operador modulo
#Esta operacion arroja el residuo de una division
#Su operador es %
B=10%3
print(B)

#Modulo en asigancion

C=20
C %=6 #Equivale a escribir C=C%6
print(C)
```

```
# Operadores de comparacion

'''
and, or, not, >, <, ==, >=, <=, !=

1. ()
2. **
3. *, /, %, not
4. +, -, and
5. >, <, ==, >=, <=, !=, or


and: si ambas expresiones tienen el valor True (cierto) da como resultado True
or: con que mÃ­nimo una expresion tenga el valor True (cierto) da como resultado True
not: cambia el valor de True (cierto) a False (falso) y viceversa 
'''

D=1
E=2
F=3

D=(D<E)and(F>E)
print(D)

D=(D>E)and(F>E)
print(D)

D=(D>E)or(F>E)
print(D)

D=(D>E)or(F<E)
print(D)

D=not(D>E)or(F>E)
print(D)
```

```
#operador "diferente" !=
A=1
G=A!=2
print(G)
G=A!=1
print(G)
print("\n")
#Operador "igualdad" ==

F=A==1
print(F)
F=A==2
print(F)
```
