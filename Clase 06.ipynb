```
#un ciclo es un cogdigo que se repite varias veces hasta que se deja 
#de cumplir una condicion

#iteracion es el numero de veces que se ha repetido el ciclo

#Ciclo while

num = int(input('Ingrese un numero par: '))

while num % 2!=0:
  num = int(input('Ingrese un numero PAR!!! '))

print('El numero par es: ',num)
```

```
#imprimiendo numeros del 1 al 5

i=1
while i <= 5:
  print(i)
  i=i+1
```

```
#Ciclo while

#ciclo donde el numero de iteraciones normalmente es conocido
#la variable i toma un valor de la lista
#en cada iteracion

#imprimir numeros del 1 al 5

for i in[1,2,3,4,5]:
  print(i)
```

```
#funcion range
#crea un arreglo de numeros
#range(Inicio, Termina, Incremento)
#range(num) el arreglo va desde cero hasta num -1

y=range(0,11,1)
print(list(y)) #para imprimir la variable con range, se debe incluir una lista

x=range(15) #es equivalente a escribir range(0,15,1)
print(list(x))
```

```
#sumatoria de los elementos en X

x = [1,2,3,4,5]

suma = 0

for i in x:
  suma = suma + i

print(suma)

#Sumando de una forma diferente
suma2 = 0
for i in range(5):
  suma2 = suma2 + x[i]

print(suma2)
```
